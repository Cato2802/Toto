```python
X = data['LSTAT']
y = data['MEDV']
X = sm.add_constant(X) # add intercept term
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())
```

* Nhận xét:

```python
# Obtain the residuals, studentized residuals and the leverages
fitted = model.fittedvalues
residuals = model.resid
studentized_residuals = model.get_influence().resid_studentized_internal
leverages = model.get_influence().hat_matrix_diag
residual_standard_error = np.sqrt(np.abs(studentized_residuals))

# Plot
plt.figure(1, figsize=(18, 20))

# Residuals
plt.subplot(321)
ax = sns.regplot(fitted, residuals, lowess=True)
plt.title('Residuals vs Fitted')
plt.xlabel('fitted values');
plt.ylabel('Studentized residual');

# Normal Q-Q plot
plt.subplot(322)
ax = stats.probplot(studentized_residuals, dist='norm', plot=plt)
plt.xlabel('Theoretical quantities')
plt.ylabel('Studentized residuals')
plt.title('Normal Q-Q')

# Scale-Location plot
plt.subplot(323)
ax = sns.regplot(fitted, residual_standard_error, lowess=True)
plt.title('Scale-Location')
plt.xlabel('fitted values')
plt.ylabel('$\sqrt{|Standardized Residuals|}$')

# Leverages
plt.subplot(324)
ax = sns.scatterplot(leverages, studentized_residuals)
plt.title('Residuals vs Leverage')
plt.xlabel('Leverage')
plt.ylabel('studentized residual')
```
